```python
# Procedure 3: Outliers
# Data Cleaning Action: Identifying and Remove the outliers
# Explanation: we identify the ouliers and decide whether to keep or drop them  .
# Calculating our first, third quantiles and then later our IQR
#

Q1 = covid.quantile(0.25)
Q3 = covid.quantile(0.75)
IQR = Q3 - Q1
IQR

# The rule of thumb is that anything not in the range of 
# (Q1 - 1.5 IQR) and (Q3 + 1.5 IQR) is an outlier, and can be removed.
#

(covid < (Q1 - 1.5 * IQR)) |(covid > (Q3 + 1.5 * IQR))

# Uncommenting the line of code below is how we removing outliers based on the IQR range and stores the result in the data frame 'electric_out'
# but for now we are going to keep them

covid_df = covid[~((covid < (Q1 - 1.5 * IQR)) |(covid > (Q3 + 1.5 * IQR))).any(axis=1)]
covid_df
```

```python
# Procedure 1: Missing values
# Data Cleaning Action: checking for missing values
# Explanation: we need to check if we have missing values 
#
# missing values
covid_df.isnull().sum()

# some missing values
```

The mean of the numerical column data is used to replace null values when the data is normally distributed. Median is used if the data comprised of outliers. Mode is used when the data having more occurences of a particular value or more frequent value.

```python
# Procedure 2: Replacing missing values
# Data Cleaning Action: replacing missing/NaN values with median
# Explanation: Instead of dropping the columns with missing values,
# we replace them with median  
#
#Finding the mean of the column having NaN
#mean_value=covid_out.mean()
  
# Replace NaNs in column S2 with the
# mean of values in the same column
#covid_out.fillna(value=mean_value, inplace=True)

#covid_df.replace(np.nan, 0, inplace=True)
#covid_df



covid_df = covid_df.fillna(covid_df.median())
covid_df
```

```python
# checking to see if the missing values have been replaced
covid_df.isnull().sum()
```

### 1.44) Consitency

```python
# Procedure 1: duplicates
# Data Cleaning Action: checking for dupicates in our dataset
# Explanation: Duplicates are data points that are repeated in your dataset. 
# These should be simply removed.
#
covid_df[covid_df.duplicated()]

# Identifying which observations are duplicates
covid_df = covid_df.drop_duplicates()

# previwing for the non duplicates

covid_df
```

### 1.45) Uniformity

```python
# Procedure 1: Standardization
# Data Cleaning Action: Fixing messy column names
# Explanation: We have column names which are in uppercase, without spaces, 
# and whitespace all around we remove this to make the column names standard
#
covid_df.columns = covid_df.columns.str.strip().str.lower()
covid_df.head()
```

```python
# inserting underscore _ to seperate the column name
#

covid_df.columns = covid_df.columns.str.replace('totaltestresults', 'total_test_results').str.replace('deathconfirmed', 'death_confirmed')
covid_df.head()
```

1. Which states had the highest number of

    a) Negative cases
    
    b) Positive cases




```python
# Which states had the highest number of Negative cases
#
highest_no = covid_df.groupby(["state"])['negative'].max()
highest_no.sort_values(ascending=False).head()
```

```python
# Which states had the highest number of Positive cases
#
highest_pos = covid_df.groupby(["state"])['positive'].max()
highest_pos.sort_values(ascending=False).head()
```

2. Which states recorded the:

    a) highest number of deaths
    
    b) lowest number of deaths


```python
# Which states recorded the highest number of deaths
#
highest_death = covid_df.groupby(["state"])['death'].max()
highest_death.sort_values(ascending=False).head(10)
```

```python
# Which states recorded the lowest number of deaths
#
lowest_death = covid_df.groupby(["state"])['death'].min()
lowest_death.sort_values(ascending=True).head()
```

```python
# Which states recorded the lowest number of recoveries
lowest_rec = covid_df.groupby(["state"])['recovered'].min()
lowest_rec.sort_values(ascending=True).head()
```
