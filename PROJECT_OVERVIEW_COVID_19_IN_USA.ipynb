```python
# Procedure 3: Outliers
# Data Cleaning Action: Remove the outliers
# Explanation: After identifying the ouliers and decided to drop them  .
# Calculating our first, third quantiles and then later our IQR
#

Q1 = covid.quantile(0.25)
Q3 = covid.quantile(0.75)
IQR = Q3 - Q1
IQR

# The rule of thumb is that anything not in the range of 
# (Q1 - 1.5 IQR) and (Q3 + 1.5 IQR) is an outlier, and can be removed.
#

(covid < (Q1 - 1.5 * IQR)) |(covid > (Q3 + 1.5 * IQR))

# The line of code below is how we remove outliers based on the IQR range and 
# stores the result in the data frame 'covid_df'


covid_df = covid[~((covid < (Q1 - 1.5 * IQR)) |(covid > (Q3 + 1.5 * IQR))).any(axis=1)]
covid_df
```

6. What is the positivity rate
